```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt


plt.style.use("ggplot")
pd.set_option('display.max_columns', 200)
```

```python
df = pd.read_csv('csv_files/coaster_db.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[[
    'coaster_name',
    'Location',
    'Status',
    'Manufacturer',
    'Height restriction',
    'Model',
    'Height',
    'Opened',
    'year_introduced',
    'latitude',
    'longitude',
    'Type_Main',
    'opening_date_clean',
    'speed_mph',
    'height_ft',
    'Inversions_clean',
    'Gforce_clean',
]]
```

```python
 # df.drop(["Opening date"], axis=1, inplace=True)
```

```python
df.shape
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
df.dtypes
```

```python
df.rename(columns={'Location':'location'}, inplace=True)
df.columns
```

```python
df.isna().sum()
```

```python
df.loc[df.duplicated()]
```

```python
df.loc[df.duplicated(subset=['coaster_name'], keep=False)]
```

```python
df.query('coaster_name == "Crystal Beach Cyclone"')
```

```python
df = df.loc[~df.duplicated(subset=['coaster_name', 'location', 'opening_date_clean'])].reset_index(drop=True).copy()
df
```

```python
df['year_introduced'].value_counts()
```

```python
ax = df['year_introduced'].value_counts().head(10).plot(kind='bar', title='Top 10 Years Coasters')
ax.set_xlabel('Years')
ax.set_ylabel('Count')
```

```python
ax = df['speed_mph'].plot(
    kind='hist',
    bins=20,
    title='Speed in mph',
)
ax.set_xlabel('Speed in mph')
ax.set_ylabel('Frequency')
```

```python
ax = df['speed_mph'].plot(
    kind='kde',
    title='Speed in mph',
)
ax.set_xlabel('Speed in mph')
ax.set_ylabel('Frequency')
```

```python
df.plot(kind='scatter', x='speed_mph', y='height_ft', title='Coaster speed vs. height')
plt.show()
```

```python
sns.scatterplot(
    data = df,
    x='speed_mph',
    y='height_ft',
    hue='year_introduced',
)
plt.show()
```

```python
sns.pairplot(
    data = df,
    vars = [
        'speed_mph',
        'height_ft',
        'Gforce_clean',
        'year_introduced',
        'Inversions_clean',
    ],
    hue='Type_Main',
)
```

```python
df_corr = df[[
    'year_introduced',
    'speed_mph',
    'height_ft',
    'Inversions_clean',
    'Gforce_clean',
]].dropna().corr()
df_corr
```

```python
sns.heatmap(
    data = df_corr,
    annot = True,
)
```

```python
df.query('location != "Other"').groupby('location')['speed_mph'].agg(['mean', 'count']).query('count >= 10').sort_values('mean', ascending=False).plot(kind='barh')
plt.show()
```
